일로아 사이트에서 캐릭터 네임 크롤링(https://iloa.gg/ranks)

5레벨 단위로 500개의 케릭터 네임을 수집

```python
### Import ###
# !pip install selenium
# !pip install webdriver_manager
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import re

### Code ###
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://iloa.gg/ranks')

range_input = driver.find_element_by_css_selector('input[type="range"]')
min_value = int(range_input.get_attribute('min'))
max_value = int(range_input.get_attribute('max'))

r = driver.find_element(By.XPATH, '//*[@id="default-range"]')
r.click()

for i in range(min_value, max_value+1, 5):
    # Move the mouse left and right over the slider to change the value.
    try:
        if i == 1415:
            action = ActionChains(driver)
            action = action.click_and_hold(range_input).move_to_element_with_offset(range_input, 1, 0)
            time.sleep(2)
            action.release().perform()
        else:
            r.send_keys(Keys.ARROW_RIGHT)
            time.sleep(2)
    except Exception as e:
        print("Error:", e)

    # Wait 1 second to check the changed slider value.
    time.sleep(1)
    wait = WebDriverWait(driver, 10)
    
    for j in range(9):
        try:
            more_button = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/main/div[2]/div[1]/div[1]/div/button/span')))
            more_button.click()
            time.sleep(3)
        except Exception as e:
            print("Error:", e)
        
    # Find and save the text content of the desired elements.
    elements = driver.find_elements_by_css_selector('span.flex.flex-col.font-semibold')
    text_content = [re.sub("\n카마인|\n루페온|\n카제로스|\n카단|\n아브렐슈드|\n니나브|\n실리안|\n아만", "", element.text) for element in elements]
    print(text_content, "\n\n\n")
    
    with open(f'iloa_{i}.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(text_content))
    time.sleep(2)
driver.quit()
```
