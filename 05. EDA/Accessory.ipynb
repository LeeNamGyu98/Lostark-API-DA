```python
import sys
sys.path.append('../package')
import lostark as lok
import pandas as pd
import pprint
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm
font_path = "C:/Users/user/Code_Files_Jupyter/nanum-all/나눔 글꼴/나눔바른고딕/NanumFontSetup_TTF_BARUNGOTHIC/NanumBarunGothic.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumBarunGothic')

pd.set_option('display.max_columns', None)
```

```python
df = lok.get_table_df('accessory_table')
```

```python
for i in range(0, len(df.columns), 20):
    df.iloc[:, i:i+20].info()
```

```python
accessories = ['목걸이', '귀걸이1', '귀걸이2', '반지1', '반지2', '팔찌', 'AS']

fig, ax = plt.subplots(4, 2, figsize=(20, 20))
idx = -1
for i in range(4):
    for j in range(2):
        idx+=1
        data = df[f'{accessories[idx]}_grade'].value_counts()
        sns.barplot(x=data.index, y=data.values, palette = 'Set2', ax=ax[i, j])
        ax[i, j].set_title(accessories[idx])
        ax[i, j].set_xlabel("Grade")
        ax[i, j].set_ylabel("Frequency")
        if idx == 4: break
plt.show()
```

```python
fig, ax = plt.subplots(4, 2, figsize=(20, 20))
idx = -1
for i in range(4):
    for j in range(2):
        idx+=1
        data = df[f'{accessories[idx]}_tier'].value_counts()
        sns.barplot(x=data.index, y=data.values, palette = 'Set2', ax=ax[i, j])
        ax[i, j].set_title(accessories[idx])
        ax[i, j].set_xlabel("Tier")
        ax[i, j].set_ylabel("Frequency")
        if idx == 4: break
plt.show()
```

```python
fig, ax = plt.subplots(4, 2, figsize=(20, 20))
idx = -1
for i in range(4):
    for j in range(2):
        idx+=1
        data = df[f'{accessories[idx]}_name'].value_counts()[:5]
        sns.barplot(x=data.index, y=data.values, palette = 'Set2', ax=ax[i, j])
        ax[i, j].set_title(accessories[idx])
        ax[i, j].set_xlabel("name")
        ax[i, j].set_ylabel("Frequency")
        if idx == 6: break
plt.show()
# 울부짖는 혼돈의, 영겁의 acc는 어디서 얻을 수 있기에 많이 착용하나? 평균 quality는? 
```

```python
acc = accessories[:5].copy()
fig, ax = plt.subplots(3, 2, figsize=(20, 20))
idx = -1
for i in range(3):
    for j in range(2):
        idx+=1
        data = df[f'{acc[idx]}_quality']
        sns.histplot(data, bins=50, kde=True, alpha=0.7, color='b', ax=ax[i, j])
        ax[i, j].set_title(accessories[idx])
        ax[i, j].set_xlabel("quality")
        ax[i, j].set_ylabel("Frequency")
        if idx == 4: break
plt.show()
```
