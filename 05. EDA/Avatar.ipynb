```python
# 아바타 등급이 레벨과 관련성이 있는가?
avatar = df.copy()
profile = lok.get_table_df('profile_table')
```

```python
parts = ['characterCode', '무기1_grade', '무기2_grade', '상의1_grade', '상의2_grade', '하의1_grade', '하의2_grade']
```

```python
df = pd.merge(avatar[parts], profile[['characterCode', 'itemMaxLevel']], on='characterCode', how='inner')
df['itemMaxLevel'] = df['itemMaxLevel'].astype('float')
df.head()
```

```python
label = {'희귀':1, '영웅':2, '전설':3, None:0}
for i in range(1, 7):
    df[df.columns[i]] = df[df.columns[i]].map(label)
```

```python
df.drop('characterCode', axis=1, inplace=True)
df['score'] = df.loc[:, :"하의2_grade"].sum(axis=1)
df = df[['itemMaxLevel', 'score']]
```

```python
cdf = df.groupby('itemMaxLevel')['score'].mean()
plt.figure(figsize=(20, 10))
plt.plot(cdf)
import numpy as np
ticks = np.arange(0, 1655, 50)
plt.xticks(ticks)
plt.title('레벨에 따른 아바타 등급 평균 점수', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(df['itemMaxLevel'], df['score'])
plt.show()
```
