```python
import pandas as pd
from pprint import pprint
```

```python
### DataBase ###
import pymysql
import json
%reload_ext sql

# JSON 파일 열기
with open('../key.json', 'r') as file:
    key = json.load(file)

# Connect to MySQL database
db = pymysql.connect(host = 'localhost', 
                     port = 3306, 
                     user='root', 
                     passwd = key['password'],
                     charset='utf8')

# Create a cursor object
cursor = db.cursor()
cursor.execute("USE lostark")
```

```python
def return_df(table_name):
    sql = f"SHOW COLUMNS FROM {table_name}"
    cursor.execute(sql)
    column_names = [column[0] for column in cursor.fetchall()]
    
    sql = f"SELECT * FROM lostark.{table_name}"
    cursor.execute(sql)
    df = pd.DataFrame(cursor.fetchall(), columns=column_names)
    return df
```

```python
df = return_df('stats_table')
```

```python
for column_name in df.columns:
    print(column_name)
    print(len(df[column_name].unique()))
    pprint(df[column_name].sort_values().unique())
    print('-'*100)
```
