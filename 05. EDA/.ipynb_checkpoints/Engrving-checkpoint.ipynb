```python
engravings = []
for i in range(1, 13):
    engravings += list(df[f'engraving{i}_name'].unique())
engravings = list(set(engravings))
engravings = [item for item in engravings if item is not None]
print(engravings)
```

```python
dfc = df.copy()
res = pd.DataFrame()
res['engraving_name'] = pd.concat([df[f'engraving{i}_name'] for i in range(1, 13, 2)], axis=0, ignore_index=True)
res['engraving_level'] = pd.concat([df[f'engraving{i}_level'] for i in range(2, 14, 2)], axis=0, ignore_index=True)
res.dropna(inplace=True)
res.groupby('engraving_name').sum().sort_values(by='engraving_level')

sns.barplot(x=res['engraving_level'], y=res['engraving_name'], orient='h')
plt.show()
```
