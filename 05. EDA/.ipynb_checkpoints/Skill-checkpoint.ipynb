```python
def preprocessing_df(df):
    res = pd.DataFrame()
    res['skill_name'] = pd.concat([df[f'skill{i}_name'] for i in range(1, 17)], ignore_index=True)
    res['skill_level'] = pd.concat([df[f'skill{i}_level'] for i in range(1, 17)], ignore_index=True)
    res['skill_cooltime'] = pd.concat([df[f'skill{i}_cooltime'] for i in range(1, 17)], ignore_index=True)
    # tripod 1, 2, 3은 구분이 필요할 것으로 추측하여 구분함
    for j in range(1, 4):
        res[f'skill_tripod{j}_name'] = pd.concat([df[f'skill{i}_tripod{j}_name'] for i in range(1, 17)], ignore_index=True)
        res[f'skill_tripod{j}_point'] = pd.concat([df[f'skill{i}_tripod{j}_point'] for i in range(1, 17)], ignore_index=True)
    res['skill_runeName'] = pd.concat([df[f'skill{i}_runeName'] for i in range(1, 17)], ignore_index=True)
    res['skill_runeGrade'] = pd.concat([df[f'skill{i}_runeGrade'] for i in range(1, 17)], ignore_index=True)
    res['skill_headAttack'] = pd.concat([df[f'skill{i}_headAttack'] for i in range(1, 17)], ignore_index=True)
    res['skill_backAttack'] = pd.concat([df[f'skill{i}_backAttack'] for i in range(1, 17)], ignore_index=True)
    res['skill_partBreak'] = pd.concat([df[f'skill{i}_partBreak'] for i in range(1, 17)], ignore_index=True)
    res['skill_stagger'] = pd.concat([df[f'skill{i}_stagger'] for i in range(1, 17)], ignore_index=True)
    res['skill_counter'] = pd.concat([df[f'skill{i}_counter'] for i in range(1, 17)], ignore_index=True)
    res['skill_마나'] = pd.concat([df[f'skill{i}_마나'] for i in range(1, 17)], ignore_index=True)
    res['skill_배터리'] = pd.concat([df[f'skill{i}_배터리'] for i in range(1, 17)], ignore_index=True)
    res['skill_버블'] = pd.concat([df[f'skill{i}_버블'] for i in range(1, 17)], ignore_index=True)
    res['skill_충격'] = pd.concat([df[f'skill{i}_충격'] for i in range(1, 17)], ignore_index=True)
    res['skill_기력'] = pd.concat([df[f'skill{i}_기력'] for i in range(1, 17)], ignore_index=True)
    res['skill_내공'] = pd.concat([df[f'skill{i}_내공'] for i in range(1, 17)], ignore_index=True)
    res['skill_영혼석'] = pd.concat([df[f'skill{i}_영혼석'] for i in range(1, 17)], ignore_index=True)
    res['skill_경직면역'] = pd.concat([df[f'skill{i}_경직면역'] for i in range(1, 17)], ignore_index=True)
    res['skill_피격면역'] = pd.concat([df[f'skill{i}_피격면역'] for i in range(1, 17)], ignore_index=True)
    res['skill_상태이상면역'] = pd.concat([df[f'skill{i}_상태이상면역'] for i in range(1, 17)], ignore_index=True)
    res = res.dropna(how='all', axis=0)
    res.reset_index(drop=True, inplace=True)
    return res
res = preprocessing_df(df)
res
```

```python
from ipywidgets import interact, widgets

className = profile_df['characterClassName'].unique()
exclued = ['암살자', '스페셜리스트', '전사(남)', '무도가(여)', '헌터(남)',
         '마법사','전사(여)', '무도가(남)', '헌터(여)']
classNames = np.setdiff1d(className, exclued)

dropdown = widgets.Dropdown(
    options=classNames,
    value = '버서커',
    description = 'Class: '
)

def show_skill_distribution_by_class(className):
    class_df = pd.merge(df, profile_df[profile_df['characterClassName'] == className]['characterCode'], 
         on='characterCode', how='inner')
    pre_class_df = preprocessing_df(class_df)
    data = pre_class_df.groupby('skill_name').size().sort_values(ascending=False)
    sns.barplot(y=data.index, x=data.values, palette='Set2', orient='h')
    plt.title(f'{className} Skill Distribution')
    plt.ylabel('Skill')
    plt.xlabel('Frequency')
    plt.show()
    
interact(show_skill_distribution_by_class, className=dropdown)
```

```python
DD_className = widgets.Dropdown(
    options=classNames,
    description = 'Class: '
)
DD_skillName = widgets.Dropdown(
    options=[],
    description = 'Skill: ' 
)
pre_class_df = pd.DataFrame()

def select_DD_skillName(className):
    global pre_class_df
    class_df = pd.merge(df, profile_df[profile_df['characterClassName'] == className]['characterCode'], 
         on='characterCode', how='inner')
    pre_class_df = preprocessing_df(class_df)
    skillNames = pre_class_df['skill_name'].unique()
    DD_skillName.options = skillNames

def show_tripod_distribution_by_skill(skillName):
    fig, ax = plt.subplots(1, 3, figsize=(17, 6))
    for i in range(3):
        tripod = pre_class_df.groupby(['skill_name', f'skill_tripod{i+1}_name']).size().reset_index()
        tripod = tripod[tripod['skill_name'] == skillName]
        sns.barplot(x=tripod[f'skill_tripod{i+1}_name'], y=tripod[0], palette='Set2', ax=ax[i])
        ax[i].set_title(f'{skillName} Tripod{i+1} Distribution')
        ax[i].set_xlabel('Tripod Name')
        ax[i].tick_params(axis='x', labelsize=12)
        ax[i].set_ylabel('Frequency')
    plt.show()

interact(select_DD_skillName, className=DD_className)
interact(show_tripod_distribution_by_skill, skillName=DD_skillName)
```

```python
DD_className = widgets.Dropdown(
    options=classNames,
    description = 'Class: '
)
DD_skillName = widgets.Dropdown(
    options=[],
    description = 'Skill: ' 
)
pre_class_df = pd.DataFrame()

def select_DD_skillName(className):
    global pre_class_df
    class_df = pd.merge(df, profile_df[profile_df['characterClassName'] == className]['characterCode'], 
         on='characterCode', how='inner')
    pre_class_df = preprocessing_df(class_df)
    skillNames = pre_class_df['skill_name'].unique()
    DD_skillName.options = skillNames

def show_rune_distribution_by_skill(skillName):
    data = pre_class_df[pre_class_df['skill_name'] == skillName]['skill_runeName'].value_counts()
    sns.barplot(x=data.index, y=data.values, palette='Set2')
    plt.title(f'{skillName} Rune Distribution')
    plt.xlabel('Rune Name')
    plt.ylabel('Frequency')
    plt.show()

interact(select_DD_skillName, className=DD_className)
interact(show_rune_distribution_by_skill, skillName=DD_skillName)
```
