```python
import sys
sys.path.append('../package')
import lostark as lok
import pandas as pd
import numpy as np
import pprint
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm
font_path = "C:/Users/user/Code_Files_Jupyter/nanum-all/나눔 글꼴/나눔바른고딕/NanumFontSetup_TTF_BARUNGOTHIC/NanumBarunGothic.ttf"
font_name = fm.FontProperties(fname=font_path).get_name()
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 설정

pd.set_option('display.max_columns', None)
```

```python
df = lok.get_table_df('special_equipment_table')
```

```python
df.head(2)
```

```python
for i in range(0, len(df.columns), 20):
    df.iloc[:, i:i+20].info()
```

```python
data = df['나침반_grade'].value_counts()
sns.barplot(x=data.index, y=data.values, palette='Set2')
plt.title('나침반 등급')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

```python
data = df['나침반_name'].value_counts()
sns.barplot(x=data.index, y=data.values, palette='Set2')
plt.title('나침반 이름')
plt.xlabel('Name')
plt.ylabel('Frequency')
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

```python
data = df['부적_grade'].value_counts()
sns.barplot(x=data.index, y=data.values, palette='Set2')
plt.title('부적 등급')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

```python
data = df['부적_name'].value_counts()
sns.barplot(x=data.index, y=data.values, palette='Set2')
plt.title('부적 이름')
plt.xlabel('Name')
plt.ylabel('Frequency')
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

```python
data = df['문장_grade'].value_counts()
sns.barplot(x=data.index, y=data.values, palette='Set2')
plt.title('문장 등급')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

```python
data = df['문장_name'].value_counts()
sns.barplot(x=data.index, y=data.values, palette='Set2')
plt.title('문장 이름')
plt.xlabel('Name')
plt.ylabel('Frequency')
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

나침반(항해) : 유물 > 전설 > 영웅 > 희귀

부적(생활) : 전설 > 희귀 > 유물 > 영웅

문장(pvp) : 유물 > 전설 > 영웅 > 희귀
