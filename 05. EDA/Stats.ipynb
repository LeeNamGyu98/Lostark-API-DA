```python
stat = ['치명', '특화', '신속', '제압', '인내', '숙련']
fig, ax = plt.subplots(2, 3, figsize=(18, 12))
idx = 0
for i in range(2):
    for j in range(3):
        sns.histplot(df[f'{stat[idx]}_값'], kde=True, ax=ax[i, j])
        ax[i, j].set_title(f'{stat[idx]}')
        ax[i, j].set_xlabel('Value')
        idx+=1
plt.show()
```

```python
stat = ['치명', '특화', '신속', '제압', '인내', '숙련']
fig, ax = plt.subplots(2, 3, figsize=(18, 12))
idx = 0
for i in range(2):
    for j in range(3):
        sns.histplot(df[f'{stat[idx]}_내실증가량'], bins=25, kde=True, ax=ax[i, j])
        ax[i, j].set_title(f'{stat[idx]} 내실증가량')
        ax[i, j].set_xlabel('Value')
        idx+=1
plt.show()
```

특정 값 이하의 값은 기본 내실 증가량으로 인해 값이 너무 많이 잡히므로 삭제함

```python
# 100초과 값은 의미 없는 값이라 판단하여 삭제
data = df['치명_치명타_적중률(%)'].value_counts()
data = data[(data.index<=100) & (data.index > 3)]
sns.lineplot(x=data.index, y=data.values)
plt.title('치명타 적중률(%)')
plt.xlabel('%')
plt.ylabel('Count')
plt.show()
```
